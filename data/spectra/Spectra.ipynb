```python
import numpy as np
%load_ext autoreload
%autoreload 2
import spectra_db_parser as sdp
```

```python
magnetic_matrix_elements, crystal_field_uk_matrix_elements, electro_matrix_elements = sdp.parse_spectra_dbf()
```

```python
p1 = lambda x: x[0] + x[1] + (x[2],)
csmme = list(map(p1, magnetic_matrix_elements))
p2 = lambda x: x[0] + (x[1],) + (x[2],) + (x[3],)
cscfme = list(map(p2, crystal_field_uk_matrix_elements))
p3 = lambda x: x[0] + x[1] + (x[2],)
cseme = list(map(p3, electro_matrix_elements))
```

```python
import csv
import numpy as np
```

```python
n = 512
randomatrix = np.random.random((n,n))
randomatrix = randomatrix + randomatrix.T
eivals = np.linalg.eigvalsh(randomatrix)
```

```python
%%timeit -n 1 -r 5
np.linalg.eigvalsh(randomatrix)
```

```python
for k, data in zip('magnetic crystal_field electro'.split(),
                   [csmme, cscfme, cseme]):
    filename = f'{k}_matrix_elements.csv'
    with open(filename, 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)
        
        # Write each sublist as a row in the CSV file
        for sublist in data:
            csvwriter.writerow(sublist)

    print(f"CSV file '{filename}' has been created.")
```
